```python
# Из сильно коррелирующих признаков оставим, наиболее коррелирующий с таргетом
ParkinsonData.drop(['MDVP:Jitter(%)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP'], axis=1, inplace=True)
ParkinsonData.drop(['MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:DDA'], axis=1, inplace=True)
```

```python
#Логистическая регрессия
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Разобъем выборку на признаки и таргет
Y = ParkinsonData.status.values
X = ParkinsonData.drop(['status'], axis=1)

# Нормализуем признаки
X = X.apply(
    lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x))
)
# Будем смотреть accuracy на кросс-валидации по 5 фолдам
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Логистическая регрессия
log = LogisticRegression()
cvsLog = cross_val_score(log, X, Y, cv=kf, scoring='accuracy').mean()
print("Logistic Regression Accuracy : {} ".format(cvsLog*100))

# kNN

def myFunc(X, target, maxKNeighbors=51, minKNeighbors=1):
    outDict = dict()
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    for i in range(minKNeighbors, maxKNeighbors + 1):
        KNC = KNeighborsClassifier(n_neighbors=i)
        cvs = cross_val_score(KNC, X, target, cv=kf, scoring='accuracy')
        outDict[i] = cvs.mean()
    return outDict

# Найдем оптимальное число соседей для KNN скользящим контролем

cvsKnn = myFunc(X, Y, maxKNeighbors=51, minKNeighbors=1)
# График для KNN

plt.figure(figsize=(10,10))
plt.plot(list(cvsKnn.keys()), list(cvsKnn.values()))
plt.ylabel('Mean accuracy')
plt.xlabel('Neighbors Number')
plt.title('Зависимость Mean Accuracy от Числа соседей')
plt.savefig('kNN.png')
plt.show()

print('Max accuracy = ', max(list(cvsKnn.values())))
print('Optimum Neighbors Number = ', list(cvsKnn.keys())[list(cvsKnn.values()).index(max(list(cvsKnn.values())))])
```

```python
# С помощью Дерева Решений найдем 5 наиболее важных признаков
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=241)
dtc.fit(X, Y)
importances = pd.Series(dtc.feature_importances_, index=X.columns.values).sort_values(ascending=False).head(5).index.values
print('Наиболее информативные признаки: ', ', '.join(importances))
```
