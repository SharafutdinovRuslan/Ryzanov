## Визуализация TSNE

```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
X,y = sklearn.datasets.load_digits()['data'],sklearn.datasets.load_digits()['target']
```

```python
%%time
X_tsne = TSNE(n_components=2,n_iter=1000,n_iter_without_progress=100,verbose=True).fit_transform(X,y)
```

```python
plt.figure(figsize=(18,10))
plt.scatter(X_tsne.transpose()[0],X_tsne.transpose()[1], c = y)
plt.colorbar()
plt.title('Визуализация tsne')
```

```python
#зафиксируем последовательность количества итераций
steps = [1,2,5,10,20,30,50,100,200,500]
#выберем 10 случайных объектов как центры кластеров
centers = X[:10]
for step in steps:
    km = KMeans(n_clusters=10,n_init=1,init=centers,max_iter=step).fit(X)
    plt.figure(figsize=(18,10))
    plt.scatter(X_tsne.transpose()[0],X_tsne.transpose()[1],c = km.predict(X))
    plt.colorbar()
    plt.title('Визуализация tsne: step = ' + str(step))
    centers = km.cluster_centers_
```
